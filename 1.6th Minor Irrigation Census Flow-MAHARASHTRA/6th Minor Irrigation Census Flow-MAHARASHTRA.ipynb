## Description

Dataset Overview:
The dataset is a part of the 6th Minor Irrigation Census conducted in Maharashtra and consists of 99,621 records. It provides comprehensive details about various minor irrigation schemes, including their type, ownership, financial costs, operational status, and water usage patterns. This dataset is crucial for understanding the current state of irrigation in Maharashtra.

Column Details:
There are 72 columns in the dataset, covering different aspects such as location details (state, district, block, village), scheme characteristics (type, owner, status), financial data (costs, subsidies), water usage patterns (pumping hours, irrigation potential), and reasons for scheme under-utilization. This makes the dataset useful for a wide range of analyses related to water resource management.

Geographical Coverage:
The dataset spans multiple districts, blocks, and villages across Maharashtra. Since minor irrigation schemes play a crucial role in rural agricultural development, this dataset provides valuable insights into the distribution and efficiency of these schemes at the local level.

Types of Irrigation Schemes:
The dataset reveals that the majority of irrigation schemes are Surface Flow Schemes, which include structures like tanks, ponds, and permanent diversions. These schemes help in rainwater harvesting and groundwater recharge, supporting irrigation in water-scarce areas.

Ownership of Schemes:
The dataset indicates that a significant number of schemes are government-owned, while a large portion is owned by individual farmers. The ownership structure is important because government-backed schemes often have better financial support, whereas privately owned schemes may rely on farmer investments or subsidies.

Financial Aspects of Schemes:
Financial details such as construction costs, machinery costs, maintenance expenses, and subsidies received are included in the dataset. These financial figures can help policymakers assess the cost-effectiveness of various schemes and plan future investments in minor irrigation.

Operational Status of Schemes:
While many schemes are operational, a notable portion is marked as not in use. However, the dataset has limited data on the exact reasons for non-functionality. Some records indicate that certain schemes have been non-operational for years, potentially due to poor maintenance, lack of funds, or changes in water availability.

Water Usage and Infrastructure:
Some records provide insights into pump operation days, irrigation potential, and culturable command areas, which can help understand how efficiently these schemes are being utilized. However, there are missing values in these fields, making it difficult to analyze overall water distribution trends accurately.

Challenges in Data Quality:
A major issue with this dataset is the presence of missing values in key columns such as lifting device details, reasons for under-utilization, and enumerator-verified data. This lack of complete information reduces the reliability of certain insights and requires further data cleaning before detailed analysis can be conducted.

Potential for Future Analysis:
Despite the data quality issues, this dataset has significant potential for identifying trends in irrigation, evaluating financial inefficiencies, and assessing scheme performance. If properly cleaned and analyzed, it can provide useful recommendations for improving water resource management, optimizing financial allocation, and ensuring better scheme utilization in Maharashtra.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("6th Minor Irrigation Census Flow-MAHARASHTRA.csv")
```

# 1. Missing Values Analysis

```python
# 1. Missing Values Analysis
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0].sort_values(ascending=False)
print("Missing Values in Each Column:")
print(missing_values)
```

# 2. Statistical Summary of Numerical Data

```python
# 2. Statistical Summary of Numerical Data
summary_stats = df.describe()
print("\nStatistical Summary:")
display(summary_stats)
```

# 3. Ownership Distribution

```python
# 3. Ownership Distribution
plt.figure(figsize=(10, 5))
ownership_distribution = df["ref_scheme_owner_name"].value_counts().head(10)
sns.barplot(x=ownership_distribution.index, y=ownership_distribution.values, palette="coolwarm")
plt.xticks(rotation=45, ha='right')
plt.xlabel("Ownership Type")
plt.ylabel("Count")
plt.title("Ownership Distribution of Irrigation Schemes")
plt.show()
```

# 4. Top Districts with Most Schemes

```python
# 4. Top Districts with Most Schemes
plt.figure(figsize=(12, 6))
top_districts = df["district_name"].value_counts().head(10)
sns.barplot(x=top_districts.index, y=top_districts.values, palette="magma")
plt.xticks(rotation=45, ha='right')
plt.xlabel("District")
plt.ylabel("Number of Schemes")
plt.title("Top 10 Districts with Most Schemes")
plt.show()
```

# 5. Trend Analysis: Commissioning Period

```python
# 5. Trend Analysis: Commissioning Period
plt.figure(figsize=(10, 5))
commissioning_trend = df["commissioning_period_display"].value_counts().sort_index()
sns.lineplot(x=commissioning_trend.index, y=commissioning_trend.values, marker="o", linestyle="-", color="b")
plt.xticks(rotation=45, ha='right')
plt.xlabel("Commissioning Period")
plt.ylabel("Number of Schemes")
plt.title("Trend of Scheme Commissioning Over Time")
plt.show()
```

# Bar chart: Distribution of scheme types

```python
# Bar chart: Distribution of scheme types
plt.figure(figsize=(6, 6))
scheme_counts = df["scheme_type_name"].value_counts().head(10)  # Top 10 scheme types
sns.barplot(x=scheme_counts.index, y=scheme_counts.values, palette="viridis")
plt.xticks(rotation=45, ha='right')
plt.xlabel("Scheme Type")
plt.ylabel("Count")
plt.title("Distribution of Scheme Types")
plt.show()
```

# Scatter plot: Construction cost vs. Culturable Command Area

```python
# Scatter plot: Construction cost vs. Culturable Command Area
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df["scheme_construction_cost"], y=df["culturable_command_area"], alpha=0.5)
plt.xlabel("Construction Cost")
plt.ylabel("Culturable Command Area")
plt.title("Construction Cost vs. Culturable Command Area")
plt.xlim(0, df["scheme_construction_cost"].quantile(0.99))  # Limit x-axis to remove extreme outliers
plt.ylim(0, df["culturable_command_area"].quantile(0.99))  # Limit y-axis similarly
plt.show()
```

# Heatmap: Correlation between numerical attributes

```python
# Heatmap: Correlation between numerical attributes
plt.figure(figsize=(12, 8))
numeric_cols = df.select_dtypes(include=["float64", "int64"]).corr()  # Selecting numerical columns
sns.heatmap(numeric_cols, cmap="coolwarm", annot=False, linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Attributes")
plt.show()
```

# 6. Boxplot of Construction Cost by Scheme Type

```python
# 6. Boxplot of Construction Cost by Scheme Type
plt.figure(figsize=(12, 6))
sns.boxplot(x="scheme_type_name", y="scheme_construction_cost", data=df)
plt.xticks(rotation=90)
plt.yscale("log")  # Log scale to handle large range of values
plt.xlabel("Scheme Type")
plt.ylabel("Construction Cost (Log Scale)")
plt.title("Construction Cost Distribution by Scheme Type")
plt.show()
```

# 7. Pie Chart: Funding Source Distribution

```python
# 7. Pie Chart: Funding Source Distribution
funding_cols = [col for col in df.columns if "finance_source_name" in col.lower()]
if funding_cols:
    plt.figure(figsize=(8, 8))
    df[funding_cols[0]].value_counts().head(6).plot.pie(autopct="%1.1f%%", cmap="Set2")
    plt.ylabel("")
    plt.title("Top 6 Funding Sources for Irrigation Schemes")
    plt.show()
else:
    print("No matching finance source column found in the dataset.")
```

# 8. Pairplot for Key Features

# 9. Violin Plot: Command Area by Scheme Type

```python
# 9. Violin Plot: Command Area by Scheme Type
plt.figure(figsize=(12, 6))
sns.violinplot(x="scheme_type_name", y="culturable_command_area", data=df)
plt.xticks(rotation=90)
plt.yscale("log")
plt.xlabel("Scheme Type")
plt.ylabel("Culturable Command Area (Log Scale)")
plt.title("Culturable Command Area Distribution by Scheme Type")
plt.show()
```

## Conclusion 

The dataset provides a comprehensive overview of minor irrigation schemes in Maharashtra, covering ownership, financial aspects, and operational status.

The majority of schemes are Surface Flow Schemes, including tanks, ponds, and permanent diversions, which are crucial for local water management.

Ownership distribution shows a mix of government-owned and individually owned schemes, with private ownership being significant in some areas.

Financial data, including construction, machinery, and maintenance costs, highlights the investment required for irrigation infrastructure.

A significant number of schemes are not in use, but the dataset lacks complete information on the reasons for their non-functionality.

Water usage data, such as pumping days and command area coverage, can help assess the efficiency of irrigation systems but contains missing values.

Data quality issues, particularly missing values in critical columns, limit the accuracy of some analyses and require cleaning for better insights.

Understanding the causes of under-utilization and addressing infrastructure gaps can improve the effectiveness of these irrigation schemes.

The dataset has potential for further analysis on financial trends, scheme efficiency, and regional disparities in irrigation coverage.

If cleaned and analyzed properly, this dataset can support policy decisions for improving minor irrigation schemes and water resource management in Maharashtra.
